## 1. Construyendo una CNN

Como se puede observar en el siguiente código, esencialmente la red convolucional que vamos a usar está formada por dos capas convoluciones bidimensionales `layer_conv_2d` seguidas de dos capas max_pooling `layer_max_pooling_2d`.

```python
from keras import layers
from keras import models
from keras.utils.vis_utils import plot_model

CNN = models.Sequential()
CNN.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
CNN.add(layers.MaxPooling2D((2, 2)))
CNN.add(layers.Conv2D(64, (3, 3), activation='relu'))
CNN.add(layers.MaxPooling2D((2, 2)))
CNN.add(layers.Conv2D(64, (3, 3), activation='relu'))

plot_model(CNN, to_file='CNN_plot.png', show_shapes=True, show_layer_names=True)
```

Veamos un resumen de la arquitectura que hemos construido:
<!--![](imgs/CNN_plot.png)-->

```python
CNN.summary()
```

Una red convolucional toma como dato de entrada un tensor de la forma `(image_height, image_width, image_channels)`. En este caso, para ajustarse a las características de las imágenes de MNIST, será `(28, 28, 1)`, ya que usaremos un solo canal (gris) en las imágenes. Observa que el tamaño del dato de entrada se pasa como argumento de la primera capa con `input_shape = (28, 28, 1)`.

Cada capa de tipo `layer_conv_2d()` y `layer_max_pooling_2d()` dan como salida un tensor 3D de forma `(height, width, channels)`. Tanto la anchura como altura del tensor tienden a disminuir a medida que avanzamos en la red. El número de canales se controla por medio del primer argumento que se le pasa a las capas `layer_conv_2d()` (que en este caso es 32 o 64).

A continuación, hemos de pasar la salida de la última capa anterior (de tamaño `(3, 3, 64)`) a una red densa clasificadora similar a las que ya hemos visto en ejemplos anteriores. Como estas capas procesan vectores, que son 1D, y la entrada es un tensor 3D, hemos de aplanar el tensor por medio de la capa `layer_flatten()`, que también proporciona Keras:

```python
CNN.add(layers.Flatten())
CNN.add(layers.Dense(64, activation='relu'))
CNN.add(layers.Dense(10, activation='softmax'))

plot_model(CNN, to_file='CNN2_plot.png', show_shapes=True, show_layer_names=True)
```

Como el objetivo es calcular una clasificación en 10 clases, la última capa es una capa densa con 10 unidades y con salida `softmax`. La red completa queda por tanto:

<!-- ![](imgs/CNN2_plot.png)-->

```python
CNN.summary()
```

Una vez definida la red, realizamos el entrenamiento de forma similar a como hicimos en el modelo simple de MNIST:


```python
from keras.datasets import mnist
from keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28, 28, 1))
test_images = test_images.astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
CNN.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
CNN.fit(train_images, train_labels, epochs=5, batch_size=64)
```

Tras el entrenamiento podemos evaluar el modelo sobre los datos de test:


```python
test_loss, test_acc = CNN.evaluate(test_images, test_labels)
```

```python
test_acc
```

Donde comprobamos que alcanzamos un accuracy de casi el 99%, un incremento relativo del 68% sobre el modelo básico visto anteriormente.

### 4.1 Carga de los datos

Como se nos explica en el README del repositorio, descargamos los paquetes que contienen los datos de entrenamiento y test y los incluímos en el directorio `data/fashion`.

Posteriormente, descargamos la función `mnist_reader` y la importamos en el mismo directorio de este archivo, para luego realizar un `import` de la función y cargar los datos haciendo referencia al directorio donde se encuentran estos y al tipo de datos que queremos recoger en cada variable.

En resumen:

- Etiquetas:

Recordemos que una vez importados los datos, debemos ajustar su forma a la correspondiente con la capa de entrada: **28x28x1**

```python
import mnist_reader

X_train, y_train = mnist_reader.load_mnist('data/fashion', kind='train')
X_test, y_test = mnist_reader.load_mnist('data/fashion', kind='t10k')

# Ajustamos la forma de los datos a la capa de entrada
X_train = X_train.reshape((60000, 28, 28, 1))
X_train = X_train.astype('float32') / 255

X_test = X_test.reshape((10000, 28, 28, 1))
X_test = X_test.astype('float32') / 255

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

### 4.2 Construcción del modelo

Pasamos a construir el mismo modelo que en la práctica, comenzamos definiendo la capa de **entrada**, convolutivas con función de activación **ReLU** y capas **MaxPooling**. Necesarias para la **extración de características de las imágenes**

```python
CNN2 = models.Sequential()
CNN2.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
CNN2.add(layers.MaxPooling2D((2, 2)))
CNN2.add(layers.Conv2D(64, (3, 3), activation='relu'))
CNN2.add(layers.MaxPooling2D((2, 2)))
CNN2.add(layers.Conv2D(64, (3, 3), activation='relu'))
```

A continuación, las capas *Fully Connected* necesarias para llevar a cabo la **clasificación**

```python
CNN2.add(layers.Flatten())
CNN2.add(layers.Dense(64, activation='relu'))
CNN2.add(layers.Dense(10, activation='softmax'))
```

Por último, el optimizador, función de pérdida y métrica de evaluación

```python
CNN2.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

### 4.3 Entrenamieno y Evaluación del modelo

Entrenamos el modelo con la función `fit`

```python
CNN2.fit(X_train, y_train, epochs=5, batch_size=64)
```

Podemos observar que, sobre el conjunto de entrenamiento, el valor de la precisión es de casi el 92%. Veamos que ocurre al evaluar el modelo con el conjunto de test.

```python
test_loss2, test_acc2 = CNN2.evaluate(X_test, y_test)
```

Obtenemos un rendimiento del 90%, nada mal. Veamos un ejemplo, vamos a imprimir por medio de `matplotlib` un ejemplo aleatorio del conjunto de test.

```python
from matplotlib import pyplot as plt

# Puede cambiar el valor si quiere probar con otros
n_example = 4003
plt.imshow(X_train[n_example], cmap='gray')
plt.show()
```

Ahora, llamamos al nuestra Red Neuronal Convolucional y pedimos que nos aporte la clase a la que pertenece la imagen. Recordemos que nos devolverá aquella cuya "probabilidad" es mayor en la función *softmax*

```python
classes = {0: "camiseta", 1: "pantalon", 2: "jersey/sudadera", 3: "vestido", 4: "abrigo", 
           5: "sandalias", 6: "camisa", 7: "zapatillas", 8: "mochila", 9: "botas"}

n_class = CNN2.predict_classes(X_train[n_example].reshape(1,28,28,1))

print("Número de clase: ", n_class)
print("El objeto es: ", classes[n_class[0]])
```
